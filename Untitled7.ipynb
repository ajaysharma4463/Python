```python
# now we will learn string formatting in python
# lets create python program with for loop

for i in range(1,13): # by using comma it automatically skip 4 spaces in next line
    print("No.{0} squarred is {1} and cube is {2}".format(i,i**2,i**3))
```

```python
for i in range(1,13): # now we will do the formating by adding gigit value with replacemt like {1:2} means letter 1 with 2 digit
    print("No.{0:2} squarred is {1:3} and cube is {2:4}".format(i,i**2,i**3))
```

```python
# now it look better 
# now we will align left side
for i in range(1,13): 
    print("No.{0:<2} squarred is {1:<3} and cube is {2:<4}".format(i,i**2,i**3))
```

```python
#by default it automatically  show in right side,
# now we canalso put centre calure by using ^
for i in range(1,13): 
    print("No.{0:<2} squarred is {1:<3} and cube is {2:^4}".format(i,i**2,i**3))
```

```python
print("pi is {0:3} value".format(22/7)) # replacemt work on integer ,not float ,15 float value
print("pi is {0:12} value".format(22/7))
print("pi is {0:16} value".format(22/7)) 
print("pi is {0:14} value".format(22/7))
print("pi is {0:12f} value".format(22/7)) #float show 6 value 
print("pi is {0:12.50f} value".format(22/7))#  python ingnore 12 and fetch 50f
print("pi is {0:50.50F} value".format(22/7))
print("pi is {0:51.50F} value".format(22/7))
print("pi is {0:52.50F} value".format(22/7))# percison is happing in after adding 10number 
print("pi is {0:62.50F} value".format(22/7))
print("pi is {0:72.50F} value".format(22/7))
print("pi is {0:72.40F} value".format(22/7))
print("pi is {0:72.60F} value".format(22/7))
```
