```python
#vertically stacking vectors

import numpy as np
v1=np.array([1,2,3,4])  # they all should be in order so that it will prododude correct result
v2=np.array([3,4,5,6])
print(v1)
print(np.vstack([v1,v2]))
print("new way to print")
print(np.vstack([v1,v2,v1,v1]))
```

```python
#horizontal stacking vector
print(np.hstack([v1,v2,v2]))
```
